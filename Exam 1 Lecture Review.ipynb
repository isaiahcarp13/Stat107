```python
import pandas as pd
```

```python
df = pd.read_csv("illini football scores.csv")
group_df = df.groupby("Opponent").agg('count').reset_index()
group_df
```

```python
df_W = df[df["Result"]=="W"]
df_L = df[df["Result"]=="L"]
```

```python
group_df_w = df_W.groupby("Opponent").agg('count').reset_index()
p = group_df_w[group_df_w["Opponent"]=="Iowa"]
p
```

```python
group_df_l = df_L.groupby("Opponent").agg('count').reset_index()
p = group_df_l[group_df_l["Opponent"]=="Iowa"]
p
```
